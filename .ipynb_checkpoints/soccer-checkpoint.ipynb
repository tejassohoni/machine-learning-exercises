```python
df1=pd.DataFrame(df, columns=['short_name', 'wage_eur', 'value_eur'])
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

wage=df1['wage_eur'].apply(value_to_float)
value=df1['value_eur'].apply(value_to_float)
df1['wage_eur']=wage
df1['value_eur']=value
df1['difference'] = df1['value_eur'] - df1['wage_eur']
df1.sort_values("value_eur", ascending=False)
df1
```

```python
import seaborn as sns
sns.set()

graph = sns.scatterplot(x='wage_eur', y='value_eur', data=df1)
graph
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool
hover = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Wage,Value)", "(@wage_eur, @value_eur)"),
    ('Name', '@short_name')
    
])
p = figure(title="soccer 2020", x_axis_label='Wage', y_axis_label='Value', plot_width=500, plot_height=500, tools=[hover])
p.circle('wage_eur', 'value_eur', size=10, source=df1)
show(p)
```
